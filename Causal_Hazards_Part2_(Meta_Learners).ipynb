**Create "treatment" variable**

```python
def create_new_treatment(row):
    if row['disaster10K'] >=1 and row['disaster10K'] < 4:
        return 'treatment_1'
    elif row['disaster10K'] >=4 and row['disaster10K'] < 7:
        return 'treatment_2'
    elif row['disaster10K'] >= 7:
        return 'treatment_3'
    else:
        return 'control'
df['treatment'] = df.apply(create_new_treatment, axis=1)
df.tail(20)
```

```python
Y = df['roa1']
T = df['treatment']
X = df[['hazard','hazardlag1','roa','dtat','lctat','cheat','chlct','invtsale','xintsale','seqat','lctlt','logat','hhi']]
#W = None
```

**S-Learner (method 1)**

```python
learner_s = BaseSRegressor(XGBRegressor(), control_name='control')
ate_s = learner_s.estimate_ate(X=X, treatment=T, y=Y, return_ci=False, bootstrap_ci=False)
ate_s
```

**T-Learner (method 2)**

```python
learner_t = BaseTRegressor(XGBRegressor(), control_name='control')
ate_t, ate_t_lb, ate_t_ub = learner_t.estimate_ate(X=X, treatment=T, y=Y)
ate_t
```
